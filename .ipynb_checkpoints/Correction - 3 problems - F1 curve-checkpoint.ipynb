```python
import numpy as np
import pandas as pd
from sklearn import svm, metrics
%pylab inline --no-import-all
```

```python
label_names = ["Nurikabe", "Ploty", "Sokoban"]
data1 = pd.io.parsers.read_csv("Data/{0}_user_time.csv".format(label_names[0]), index_col=0)
data2 = pd.io.parsers.read_csv("Data/{0}_user_time.csv".format(label_names[1]), index_col=0)
data3 = pd.io.parsers.read_csv("Data/{0}_user_time.csv".format(label_names[2]), index_col=0)
data = data1.join(data2, how="inner").join(data3, how="inner")

corr = data.corr(method="spearman")
corr.unstack().hist()
```

```python
def run_test(error_rate = 0.1):
    labels = np.array([0] * data1.columns.size + [1] * data2.columns.size + [2] * data3.columns.size)
    errors = np.random.rand(labels.size) < error_rate
    for i in range(len(errors)):
    if errors[i]:
        labels[i] = (labels[i] + (randint(0,1)*2 - 1) ) % len(label_names)
    
    clf = svm.SVC(kernel="linear")
    clf.fit(corr, labels)
    predict = np.array(clf.predict(corr))
    
    found_errors = predict != labels
    f1 = metrics.f1_score(errors, found_errors, labels=[0,1], average=None)
    recall = metrics.recall_score(errors, found_errors, labels=[0,1], average=None)
    precision = metrics.precision_score(errors, found_errors, labels=[0,1], average=None)
    accuracy = metrics.accuracy_score(errors, found_errors)
    if np.sum(errors) == 0:
        f1[1] = 1.
        recall[1] = 1.
        precision[1] = 1.
    return np.hstack((f1,recall,precision,accuracy))
```

```python
returnd_values = 7
iterations = 100
points = np.arange(0, 1, 0.01)
values = np.empty((returnd_values, len(points)))
for i, error_rate in enumerate(points):
    results = np.empty((returnd_values, iterations))
    for j in range(iterations):
        results[:, j] = run_test(error_rate=error_rate)
    values[:, i] = np.mean(results, axis=1)
```

```python
plt.plot(points, values[0], "b-")
plt.plot(points, values[1], "r-")
plt.plot(points, values[2], "b--")
plt.plot(points, values[3], "r--")
plt.plot(points, values[4], "b:")
plt.plot(points, values[5], "r:")
plt.plot(points, values[6], "g-")
plt.savefig("-".join(label_names) + ".png")
```
