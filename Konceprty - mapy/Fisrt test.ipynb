```python
def all_perms(elements):
    if len(elements) <=1:
        yield elements
    else:
        for perm in all_perms(elements[1:]):
            for i in range(len(elements)):
                yield perm[:i] + elements[0:1] + perm[i:]

def swapSol(sol1, sol2, clusterNumber):
    best_pr = 0.0
    for p in all_perms(range(clusterNumber)):
        hits = 0.0
        for i in range(clusterNumber):
            hits += sum((sol1==i) & (sol2==p[i]))
        pr = hits/sol1.size
        if pr > best_pr:
            best_pr = pr
            best = p
    
    sol3 = np.copy(sol2)
    for i in range(clusterNumber):
        sol3[sol2 == best[i]] = i
    return sol3
```

```python
# -*- coding: utf-8 -*-
def clusterSpearmanSC(cor, sol, clusterNumber = 2, SCtype = 1, KMiter=20, kcut=0, plot = False, mutual=False):
    print "Clustering ..."
    SC = spectralclustering.Spectral_Clusterer(np.copy(cor))
    if kcut>0:
        if mutual:
            SC.kNearestMutual(kcut)
        else:
            SC.kNearest(kcut)
    
    solSC = SC.run(clusterNumber, SCtype, KMiter)
    solSC = swapSol(sol,solSC,clusterNumber)        
        
    
    return np.sum(sol==solSC)*1.0/len(sol), solSC, SC
```

```python
anss = answers.from_csv("data/geography.answer.csv", "data/geography.answer_options.csv")
```

```python
#ans = anss.ix[:100000]
ans = answers.first_answers(anss, "user")

pls = list(set(ans["place_asked"]))
usrs = list(set(ans["user"]))
results = pd.DataFrame(index=usrs, columns=pls)
times = pd.DataFrame(index=usrs, columns=pls)

for _, a in ans.iterrows():
    p = a["place_asked"]
    u = a["user"]
    if np.isnan(results.ix[u, p]):
        results.ix[u, p] = 1 * (a["place_asked"] == a["place_answered"])
        if a["number_of_options"] > 0:
            results.ix[u, p] -= .1 / a["number_of_options"]
        times.ix[u, p] = 1 * (a["response_time"])
        
results.save("results.df")
times.save("times.df")
```

```python
results_filtered = results.ix[results.count(1)>20, results.count()>50]
results_filtered = results_filtered.astype(float)
corr = results_filtered.corr(method="spearman")
corr.to_pickle("correlations.pd")
```

```python
ps = places.from_csv("data/geography.place.csv", "data/geography.placerelation.csv","data/geography.placerelation_related_places.csv")

continents = {
        232: "australie",
        231: "asie",
        230: "evropa",
        229: "severni amerika",
        228: "jizni amerika",
        227: "afrika",
        225: "svet",
}
continents = {225: "svet",}

def get_continent(place):
    for id, t in place["relations"]:
        if t=='is_on_map' and id in continents.keys():
            return id

def in_selection(place):
    for id, t in place["relations"]:
        if t=='is_on_map' and id in continents.keys():
            return True
    return False

selected_places = []
for _, p in ps.iterrows():
    if in_selection(p) and p["type"]==1:
        #print p["name"], p["relations"],  in_selection(p), p["type"]==1
        selected_places.append(p["id"])
```

```python
#res = times.astype(float)
res = results.astype(float)
sel = (list(set(res.columns) & set(selected_places)))
res = res.loc[:,sel]
corr = res.corr(method="spearman")
corr = corr.dropna(axis=0, how='all', thresh=None, subset=None)
corr = corr.dropna(axis=1, how='all', thresh=None, subset=None)
corr = corr.fillna(0)
solution = np.zeros(len(corr.index))
```

```python
pr,labels,SC = spectralclustering.clusterSpearmanSC(corr, solution, clusterNumber = 4, KMiter = 50, kcut = corr.shape[0]/2, SCtype = 2, plot = False, mutual = False)
```
