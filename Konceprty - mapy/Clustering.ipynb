```python
from proso.geography import places, answers, dfutil
import sys
sys.path.append("..")
from Algoritmy import spectralclustering
import numpy as np
import pandas as pd
%pylab inline --no-import-all
```

```python
answers_all = answers.from_csv("data/geography.answer.csv", "data/geography.answer_options.csv")
places_all = places.from_csv("data/geography.place.csv", "data/geography.placerelation.csv","data/geography.placerelation_related_places.csv")
```

```python
continents = {
        232: "australie",
        231: "asie",
        230: "evropa",
        229: "severni amerika",
        228: "jizni amerika",
        227: "afrika",
        225: "svet",
}

selection = [230]

def get_continent(place):
    for id, t in place["relations"]:
        if t=='is_on_map' and id in continents.keys():
            return id

def in_selection(place):
    for id, t in place["relations"]:
        if t=='is_on_map' and id in selection:
            return True
    return False

places_selected = []
for _, place in places_all.iterrows():
    if in_selection(place) and place["type"]==1:
        places_selected.append(place["id"])
        
", ".join([str(id) for id in places_selected])
```

```python
ans = answers_all[answers_all.place_asked.isin(places_selected)]
ans = answers.first_answers(ans, "user")
len(ans)
```

```python
def compute_value(a):
    return a["place_asked"] == a["place_answered"]

values = ans.groupby(["place_asked", "user"]).apply(compute_value)
```
