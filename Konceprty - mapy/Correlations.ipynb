```python
maps = {}
for id, name in continents.items():
    maps[name + "-staty"] = []
    maps[name + "-mesta"] = []
    
for p, place in places_all.iterrows():
    for id,t in place["relations"]:
        if t=='is_on_map' and id != 225:
            if place["type"] == 1:
                maps[continents[id] + "-staty"].append(p)
            else:
                maps[continents[id] + "-mesta"].append(p)
                
for name, places in maps.items():
    if len(places) == 0:
        del maps[name]
```

```python
answers2 = answers_all[answers_all.place_asked.isin(places)]
answers = answers.first_answers(answers2, ["user"])
len(answers)
```

```python

users = list(set(answers["user"]))
results = pd.DataFrame(index=users, columns=places)
results_raw = pd.DataFrame(index=users, columns=places)

for _, answer in answers.iterrows():
    p = answer["place_asked"]
    u = answer["user"]
    results.ix[u, p] = 1 * (answer["place_asked"] == answer["place_answered"])
    if answer["number_of_options"] > 0:
        results.ix[u, p] -= .1 / answer["number_of_options"]
    results_raw.ix[u, p] = 1 * (answer["place_asked"] == answer["place_answered"])
```

```python
#results_filtered = results.ix[results.count(1)>50, results.count()>10]
results_filtered = results.astype(float)
results_raw_filtered = results_raw.astype(float)
corr = results_filtered.corr(method="spearman")
corr_raw = results_raw_filtered.corr(method="spearman")
#corr = corr.dropna(axis=0, how='all', thresh=None, subset=None)
#corr = corr.dropna(axis=1, how='all', thresh=None, subset=None)
#corr = corr.fillna(0)
solution = np.zeros(len(corr.index))
```
