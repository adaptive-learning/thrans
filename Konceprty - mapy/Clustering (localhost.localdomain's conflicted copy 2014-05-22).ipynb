```python
from proso.geography import places, answers
import sys
sys.path.append("..")
from Algoritmy import spectralclustering
import numpy as np
import pandas as pd
%pylab inline --no-import-all
```

```python
answers_all = answers.from_csv("data/geography.answer.csv", "data/geography.answer_options.csv")
places_all = places.from_csv("data/geography.place.csv", "data/geography.placerelation.csv","data/geography.placerelation_related_places.csv")
```

```python
continents = {
        232: "australie",
        231: "asie",
        230: "evropa",
        229: "severni amerika",
        228: "jizni amerika",
        227: "afrika",
        225: "svet",
}

selection = [225]

def get_continent(place):
    for id, t in place["relations"]:
        if t=='is_on_map' and id in continents.keys():
            return id

def in_selection(place, selection):
    for id, t in place["relations"]:
        if t=='is_on_map' and id in selection:
            return True
    return False

places_selected = []
for _, place in places_all.iterrows():
    if in_selection(place, selection) and place["type"]==1:
        places_selected.append(place["id"])
        
", ".join([str(id) for id in places_selected])
```

```python
answers = answers_all[answers_all.place_asked.isin(places_selected)]
len(answers)
```

```python
places = places_selected
users = list(set(answers["user"]))
results = pd.DataFrame(index=users, columns=places)
times = pd.DataFrame(index=users, columns=places)

for _, answer in answers.iterrows():
    p = answer["place_asked"]
    u = answer["user"]
    if np.isnan(results.ix[u, p]):
        results.ix[u, p] = 1 * (answer["place_asked"] == answer["place_answered"])
        if answer["number_of_options"] > 0:
            results.ix[u, p] -= .1 / answer["number_of_options"]
        times.ix[u, p] = 1 * (answer["response_time"])
```

```python
results_filtered = results.ix[results.count(1)>50, results.count()>10]
results_filtered = results_filtered.astype(float)
corr = results_filtered.corr(method="spearman")
corr = corr.dropna(axis=0, how='all', thresh=None, subset=None)
corr = corr.dropna(axis=1, how='all', thresh=None, subset=None)
corr = corr.fillna(0)
solution = np.zeros(len(corr.index))
```

```python
pr,labels,SC = spectralclustering.clusterSpearmanSC(corr, solution, clusterNumber = 4, KMiter = 50, kcut = corr.shape[0]/2, SCtype = 3, plot = False, mutual = False)
```

```python
clusters = {}
for continent in selection:
    sel = []
    for _, place in places_all.iterrows():
        if in_selection(place, [continent]) and place["type"]==1:
            sel.append(place["id"])
    clusters[continents[continent]] = sel
```

```python
import json
with open("clusters_continents.json", "w") as f:
    json.dump(clusters, f)
```
