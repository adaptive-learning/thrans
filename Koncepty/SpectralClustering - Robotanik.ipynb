```python
import sys
sys.path.append("..")
from Algoritmy import spectralclustering
import numpy as np
import pandas as pd
%pylab inline --no-import-all
```

```python
# -*- coding: utf-8 -*-
import codecs
names = {}
f = codecs.open("Data/{0}_mapping.txt".format("Robotanik"))
for line in f.readlines():
    key, value = line.split(";")
    names[key] = value
```

```python
data = pd.io.parsers.read_csv("Data/{0}_user_time_new.csv".format("Robotanik"), index_col=0)
data = data.ix[data.count(1)>10,(data.count()>5)]

corr = data.corr(method="spearman")
corr.unstack().hist()
```

```python
def all_perms(elements):
    if len(elements) <=1:
        yield elements
    else:
        for perm in all_perms(elements[1:]):
            for i in range(len(elements)):
                yield perm[:i] + elements[0:1] + perm[i:]

def swapSol(sol1, sol2, clusterNumber):
    best_pr = 0.0
    for p in all_perms(range(clusterNumber)):
        hits = 0.0
        for i in range(clusterNumber):
            hits += sum((sol1==i) & (sol2==p[i]))
        pr = hits/sol1.size
        if pr > best_pr:
            best_pr = pr
            best = p
    
    sol3 = np.copy(sol2)
    for i in range(clusterNumber):
        sol3[sol2 == best[i]] = i
    return sol3
```

```python
# -*- coding: utf-8 -*-
def clusterSpearmanSC(cor, sol, clusterNumber = 2, SCtype = 1, KMiter=20, kcut=0, plot = False, mutual=False):
    print "Clustering ..."
    SC = spectralclustering.Spectral_Clusterer(np.copy(cor))
    if kcut>0:
        if mutual:
            SC.kNearestMutual(kcut)
        else:
            SC.kNearest(kcut)
    
    solSC = SC.run(clusterNumber, SCtype, KMiter)
    solSC = swapSol(sol,solSC,clusterNumber)
    
    
    if plot:
        plt.figure()
        plt.plot(SC.eig_vect[solSC==0,1],SC.eig_vect[solSC==0,2],"ro")
        plt.plot(SC.eig_vect[solSC==1,1],SC.eig_vect[solSC==1,2],"go")
        plt.plot(SC.eig_vect[solSC==2,1],SC.eig_vect[solSC==2,2],"bo")
        plt.plot(SC.eig_vect[solSC==3,1],SC.eig_vect[solSC==3,2],"yo")
        for n in range(len(data.columns)):
            #plt.text(SC.eig_vect[n,1],SC.eig_vect[n,2], data.columns[n])
            if str(data.columns[n]) in names.keys():
                x = (names[str(data.columns[n])]).encode()
                plt.text(SC.eig_vect[n,1],SC.eig_vect[n,2], x)
    
    return np.sum(sol==solSC)*1.0/len(sol), solSC, SC
```

```python
solution = np.array([0] * data.columns.size)
```

```python
# -*- coding: utf-8 -*-
 
pr,labels,SC = clusterSpearmanSC(corr, solution, clusterNumber = 4, KMiter = 50, kcut = corr.shape[0]/2, SCtype = 2, plot = True, mutual = False)
plt.savefig("Robotanik - names.png")
```

```python
stats = pd.io.parsers.read_csv("Data/{0}_stats.csv".format("Robotanik"), index_col=0)
stats2 = {}
cos = ["median_casu","avg_casu","med_casu","avg_tahu","solved","attempts","min_solving_time","max_solving_time",
       "1PM_constant","3PM_constant","3PM_slope","3PM_randomness"]
for co in cos:
    print co
    stats2[co] = x = {}
    y = []
    for n in range(len(data.columns)):
        if int(data.columns[n]) in stats[co].keys():
            time = stats[co][int(data.columns[n])]
            if type(time)==str:
                time = time.split(":")
                time = int(time[0])*60 + int(time[1])
            x[data.columns[n]] = time
            y.append(time)
    stats2[co]["data"] = y
    stats2[co]["sorted"] = sorted(y)
    stats2[co]["max"] = max(y)
    stats2[co]["min"] = min(min(y),0)
```

```python
def clusterSpearmanSC2(cor, sol, clusterNumber = 2, SCtype = 1, KMiter=20, kcut=0, plot = False, mutual=False, co = ""):
    print "Clustering ..."
    SC = spectralclustering.Spectral_Clusterer(np.copy(cor))
    if kcut>0:
        if mutual:
            SC.kNearestMutual(kcut)
        else:
            SC.kNearest(kcut)
    
    solSC = SC.run(clusterNumber, SCtype, KMiter)
    solSC = swapSol(sol,solSC,clusterNumber)
    
    
    if plot:
        plt.figure()
        for n in range(len(data.columns)):
            name = data.columns[n]
            if name in stats2[co].keys():
                v = 1 - (1.*stats2[co][name]-stats2[co]["min"])/ (stats2[co]["max"]-stats2[co]["min"])
                #v = 1 - stats2[co]["sorted"].index(stats2[co][name]) / float(len(stats2[co]["sorted"]))
                plt.plot(SC.eig_vect[n,1],SC.eig_vect[n,2], "o", color=(v,v,v, 1))
    
    return np.sum(sol==solSC)*1.0/len(sol), solSC, SC
```

```python
for co in cos:
    plt.rcParams['figure.figsize'] = 20, 20
    pr,labels,SC = clusterSpearmanSC2(corr, solution, clusterNumber = 4, KMiter = 50, kcut = corr.shape[0]/2, SCtype = 2, plot = True, mutual = False, co =co)
    plt.savefig("Robotanik - {0}.png".format(co))
```
